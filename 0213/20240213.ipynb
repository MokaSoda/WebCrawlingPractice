```python
import datetime
import random
import re
from sqlalchemy.sql import bindparam

# handler = conn.connect()
random.seed(None)
header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
handler = conn.connect()
def store(title, content):
    query = text(
        """
        insert into pages (title, content)
        values (:x, :y)
        """)
    query = query.bindparams(
        bindparam('x', title),
        bindparam('y', content)
    )
    
    handler.execute(query)
    handler.commit()
    
async def getLinks(session: aiohttp.ClientSession, articleUrl):
    # useragent 


    result = await session.get(url = 'https://en.wikipedia.org'+ articleUrl)
    html = await result.content.read()
    bs = BeautifulSoup(html, 'lxml')
    title = bs.select_one('h1').get_text()
    content = bs.select_one('div#mw-content-text p').get_text().strip()
    store(title, content)
    return bs.select_one('div#bodyContent').findAll('a', href=re.compile('^(\/wiki\/)((?!:).)*$'))

async with aiohttp.ClientSession(headers=header) as session:
    links = await getLinks(session, '/wiki/Kevin_Bacon')
    cnt = 0
    try:
        while len(links) > 0:
            newArticle = links[random.randint(0, len(links)-1)].attrs['href']
            print(newArticle)
            links = await getLinks(session, newArticle)
            cnt += 1
            if cnt > 10:
                break
    finally:
        handler.close()
    
    
        
        
```
